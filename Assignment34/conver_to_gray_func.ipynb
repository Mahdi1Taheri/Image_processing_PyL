```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rgb2gray(img):
    gary_img = 0.299 * img[:,:,2] + 0.587 * img[:,:,1] + 0.114 * img[:,:,0]
    return gary_img

img = cv2.imread('input/rubik.png')
img2 = rgb2gray(img)
plt.imshow(img2,cmap='gray')
plt.show()
```
